Goal is to produce maps that can be printed off to play with projections

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
# def equidistant_lines(ax):
#     ax.

import cartopy.geodesic
geod = cartopy.geodesic.Geodesic()

start = [-60, 40]
end = [60, 40]


import numpy as np


npts = 50
xs = np.linspace(start[0], end[0], npts)
ys = np.linspace(start[1], end[1], npts)
points = np.vstack([xs, ys]).T
points.shape



import shapely.geometry as sgeom

gc = sgeom.LineString(points[(0, -1), :])
# rhumb = sgeom.LineString(points)


def decorate(ax):
    ax.gridlines()
    ax.coastlines()
    ax.tissot(alpha=0.35, rad_km=800, facecolor='orange')
    
    ax.add_geometries([gc], ccrs.Geodetic(), edgecolor='red', facecolor='none', linewidth=2)
#     ax.add_geometries([rhumb], ccrs.Geodetic(), edgecolor='blue', facecolor='none', linewidth=2)
# gc
```

```python
plt.figure(figsize=(12, 12))
ax = plt.axes(projection=ccrs.Robinson())
decorate(ax)
```

```python
projections = [ccrs.Robinson(), ccrs.PlateCarree(), ccrs.Mercator(), ccrs.AlbersEqualArea(), ccrs.Sinusoidal(),
               ccrs.Miller(), ccrs.InterruptedGoodeHomolosine(), Ortho(),
               ccrs.LambertAzimuthalEqualArea(), ccrs.NorthPolarStereo(), ccrs.Mercator(), ccrs.Gnomonic()]
```

```python
from matplotlib import pyplot as plot
from matplotlib.backends.backend_pdf import PdfPages
 
# The PDF document
pdf_pages = PdfPages('projections.pdf')

proj_per_page = 3
n_pages = max([len(projections) // proj_per_page, 1])
for i in range(n_pages):
    # Create a figure instance (ie. a new page)
    fig = plot.figure(figsize=(8.27, 11.69), dpi=100)

    page_projections = projections[proj_per_page*i:(proj_per_page*i + proj_per_page)]
    for j, proj in enumerate(page_projections):
        # Plot whatever you wish to plot
        ax = fig.add_subplot(proj_per_page, 1, j + 1, projection=proj)
        decorate(ax)

    # Done with the page
    pdf_pages.savefig(fig)

# Write the PDF document to the disk
pdf_pages.close()
```
